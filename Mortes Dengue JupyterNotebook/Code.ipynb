```python
#AQUI ESTOU IMPORTANDO TODAS AS BIBLIOTECAS QUE USAREI DURANTE O PROGRAMA
#Para plotar os gráficos no jupyter
%matplotlib inline

import pandas as pd
from pandas import *
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('tableau-colorblind10')  #print(plt.style.available) para ver os styles disponíveis

# style diz respeito ao estilo dos gráficos
```

```python
#Comando abaixo para acessar o arquivo csv escolhido
dg = pd.read_csv('Dengue 2016 estados.csv', sep=';', engine='python')
```

```python
# Exibindo o conteúdo do arquivo
dg
```

```python
#Com base na tabela, vimos que a idade de pessoas que mais morrem são as de 80 anos para cima,
# Defini uma variável para pegar só a coluna de 80 anos com a coluna das regiôes e deixar em ordem crescente
# Criei uma tabela do tipo 'barh' que é a barra horizontal com a cor vermelho escuro
mortes80 = dg[['Região/Unidade da Federação', '80 anos e mais']].set_index('Região/Unidade da Federação').sort_values('80 anos e mais', ascending=False)
mortes80.plot(kind='barh', figsize=(12,7), grid=False, color='darkred', legend = True)

#Os comandos abaixos são para definir o título do gráfico e os textos nos eixos 'y' e 'x'
plt.title('Número de mortes com pessoas de mais de 80 anos')
plt.xlabel('Número de óbitos')
plt.show()
```

<h1>Local com mais mortes é na Região Sudeste do país</h1>

Pode se pensar que é a que tem maior índice de morte porque essa região é a região mais populosa do país.

Na linha abaixo vejo quantas vezes um número de mortes se repete na coluna de 80 anos e mais

```python
dg['80 anos e mais'].value_counts().plot(kind='bar', figsize=(11,5), grid = False, rot=0, color='green')

plt.title('Número de mortes')
plt.xlabel('Quantas mortes')
plt.ylabel('Vezes que o número se repete')
plt.show()
```

```python
#Agora estou criando um gráfico de pizza mostrando todo o total de mortes e o número que representa só a coluna de 80 anos +

x_list = [874, 234]    # 874 é o total menos o 234
labels_list = ['Total restante','80 anos e mais',]
total = sum(x_list)
explode = (0, 0.1)  # vai tirar uma fatia no gráfico

plt.axis('equal')   # diz para o gráfico ficar redondo
plt.pie(x_list, explode = explode, labels = labels_list, autopct=lambda p: '{:.0f}'.format(p * total / 100),
       shadow=True, startangle = 5)

plt.title('Comparação do total de mortes')
plt.show
```
